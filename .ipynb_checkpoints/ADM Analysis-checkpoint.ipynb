# Ft Lewis Intervention Analysis

```python
import pandas as pd
import numpy
from IPython.display import display, HTML 
%matplotlib inline
```

## Global Variable Declarations and Statistical Loader Functions

#### Global Variable Declarations include directory pointers, base simulation numbers, intervention ID's of note, experimental parameters for analysis, and control run data

```python
#Directory and statistic locations
admStats = "/active_duty_military_DetailStats.csv"
cwopStats = "/civilian_workers_on_post_DetailStats.csv"
admFile = "active_duty_military"
cwopFile = "civilian_workers_on_post"

#Population sizes
admCount = 27663
cwopCount = 15912

#Seed day
seedDay= 1

#Interventions of note
strongImpStrongEffect = "ve70vtd17allonpost/ate87atl30apl30/sq90sqg25sqtd10sql28"
strongImpWeakEffect = "ve30vtd17allonpost/ate25atl30apl30/sq90sqg25sqtd10sql28"
weakImpStrongEffect = "ve70vtd38active/ate87atl10apl30/sq90sqg25sqtd10sql7"
weakImpWeakEffect = "ve30vtd38active/ate25atl10apl30/sq90sqg25sqtd10sql7"

#Variables of note
controlled = ['v','vtd','atdr','atti','attd','atl','apl','sq','sqg','sqtd','sql']
unused = ['v','atdr','atti','attd','sq','sqg','sqtd']
natural = ['ve','ate']
experimental = filter(lambda x:x not in unused,controlled)
epistats = ['attackRate','peakDay','peakNumber']
ADMControlStats = getExperiment('Control'+ admStats)['means']
CWOPControlStats = getExperiment('Control'+ cwopStats)['means']
```

#### Loader, access, and formating functions to pull raw statistics from GetSlice.py output files

```python
#Pulls attack rate from dataframe
def getAR(meanData,cell):
    return meanData.ix[cell]['attackRate']

#Loads curve for target cell
def loadSingleCell(folder,subpop,cell,label):
    fileIn = folder + '/' + subpop + '_'*(len(subpop) > 0) + 'MeanPlots.csv'
    data = pd.read_csv(fileIn, index_col='index')
    toReturn = data.ix[cell]
    toReturn.name = label
    return toReturn

#Loads whole experiment from GetSlice.py output
def getExperiment(experimentCSV):
    data = pd.read_csv(experimentCSV, skipinitialspace=True, index_col='directory')
    means = data[data['iteration']=='Mean'].replace(-1,NaN).replace('Mean',NaN).dropna(axis=1)
    iterations = data[data['iteration']!='Mean'].replace(-1,NaN).replace('Mean',NaN).dropna(axis=1)    
    return {'data':data,'means':means,'iterations':iterations}

#Automatically pulls seed number for experiment, allows pairing with proper control run
def getSeedNumber(expDir):
    return str(loadSingleCell(expDir,admFile,strongImpStrongEffect,'null')[seedDay])

#Percentage text formatting
def percent(number):
    return "{0:.3f}%".format(number * 100)
```

## Statistical AnalysisFunctions

#### getCorrelations: using cell iteration data rather than resultant mean data, function derives, outputs, and returns correlations between experimentally controlled parameters and resultant peak day, attack rate, and peak height.

```python
def getCorrelations(iterations,natural,experimental):
    attackNatural = iterations[natural].corrwith(iterations['attackRate'])
    numberNatural = iterations[natural].corrwith(iterations['peakNumber'])
    dayNatural = iterations[natural].corrwith(iterations['peakDay'])
    attackExperimental = iterations[experimental].corrwith(iterations['attackRate'])
    numberExperimental = iterations[experimental].corrwith(iterations['peakNumber'])
    dayExperimental = iterations[experimental].corrwith(iterations['peakDay'])
    
    naturalStats = pd.concat([attackNatural, dayNatural, numberNatural], join='outer', axis = 1, names=['attackRate','peakDay','peakNumber'])
    experimentalStats = pd.concat([attackExperimental, dayExperimental, numberExperimental], join='outer', axis = 1, names=['attackRate','peakDay','peakNumber'])
    naturalStats.columns = experimentalStats.columns = ['attackRate','peakDay','peakNumber']
    
    print "Experimental Variable v. Epistat Correlation coefficients:\n"
    print "Pharmaceutical Effectiveness\n",naturalStats
    print "\nIntervention Sequence\n", experimentalStats,'\n\n'
    return {'naturalStats':naturalStats,'experimentalStats':experimentalStats}
```

#### getMaxima: fuction compares experimental cell mean stats to identify the most and least efffective interventions, outputs the intervention parameters of each, and passes on their IDs for plotting. 

```python
def getMaxima(means):
    weakest = means.ix[means['attackRate'].idxmax()]
    strongest = means.ix[means['attackRate'].idxmin()]
    weakRef = weakest.name; strongRef = strongest.name
    weakest.name = 'Weakest Intervention'; strongest.name = 'Strongest Intervention'
    maxMin = pd.concat([weakest, strongest], axis= 1)
    
    print "Attack Rate Extrema by Intervention Parameters\n\n",maxMin,'\n\n'
    return {'weakRef':weakRef,'strongRef':strongRef,'maxMin':maxMin}
```

#### getCurves: Function takes most/ least effective intervention ID's for a given experiment and outputs mean attack rate by day and mean cumulative attack rate plots for each versus their respective control experiment

```python
def getCurves(expDir,weakRef,strongRef,admFile,end,control):
    admBestCurve = loadSingleCell(expDir,admFile,strongRef,'Strongest')[0:end]
    admWorstCurve = loadSingleCell(expDir,admFile,weakRef,'Weakest')[0:end]
    admControlCurve = loadSingleCell('Control','',control,'Control')[0:end]
    
    admCurves = pd.DataFrame(zip(admBestCurve,admWorstCurve,admControlCurve), columns=[admBestCurve.name,admWorstCurve.name,admControlCurve.name], index=range(end))
    cumsumCurves = admCurves.cumsum()
    
    admCurves.plot()
    cumsumCurves.plot()
    return admCurves
```

#### getRankedInterventions: Function analyzes high/ low drug efficacies vs most/ least aggresive logistical implementations (rapid implementation, long administration) as compared to actual most/ least effective implementations and the experimental control run

```python
def getRankedInterventions(ADMMeanStats, CWOPMeanStats, best, worst, control):
    SISE_ADM_AttackRate = getAR(ADMMeanStats, strongImpStrongEffect)
    SISE_CWOP_AttackRate = getAR(CWOPMeanStats, strongImpStrongEffect)
    SIWE_ADM_AttackRate = getAR(ADMMeanStats, strongImpWeakEffect)
    SIWE_CWOP_AttackRate = getAR(CWOPMeanStats, strongImpWeakEffect)
    WISE_ADM_AttackRate = getAR(ADMMeanStats, weakImpStrongEffect)
    WISE_CWOP_AttackRate = getAR(CWOPMeanStats, weakImpStrongEffect)
    WIWE_ADM_AttackRate = getAR(ADMMeanStats, weakImpWeakEffect)
    WIWE_CWOP_AttackRate = getAR(CWOPMeanStats, weakImpWeakEffect)
    
    Best_ADM_AttackRate = getAR(ADMMeanStats, best)
    Worst_ADM_AttackRate = getAR(ADMMeanStats, worst)
    Control_ADM_AttackRate = getAR(ADMControlStats, control)
    Best_CWOP_AttackRate = getAR(CWOPMeanStats, best)
    Worst_CWOP_AttackRate = getAR(CWOPMeanStats, worst)
    Control_CWOP_AttackRate = getAR(CWOPControlStats, control)
    
    SISE_ADM_Percent = float(SISE_ADM_AttackRate) / admCount
    SISE_CWOP_Percent = float(SISE_CWOP_AttackRate) / cwopCount
    SIWE_ADM_Percent = float(SIWE_ADM_AttackRate) / admCount
    SIWE_CWOP_Percent = float(SIWE_CWOP_AttackRate) / cwopCount
    WISE_ADM_Percent = float(WISE_ADM_AttackRate) / admCount
    WISE_CWOP_Percent = float(WISE_CWOP_AttackRate) / cwopCount
    WIWE_ADM_Percent = float(WIWE_ADM_AttackRate) / admCount
    WIWE_CWOP_Percent = float(WIWE_CWOP_AttackRate) / cwopCount
    
    Best_ADM_Percent = float(Best_ADM_AttackRate) / admCount
    Worst_ADM_Percent = float(Worst_ADM_AttackRate) / admCount
    Control_ADM_AttackRate = float(Control_ADM_AttackRate) / admCount
    Best_CWOP_Percent = float(Best_CWOP_AttackRate) / cwopCount
    Worst_CWOP_Percent = float(Worst_CWOP_AttackRate) / cwopCount
    Control_CWOP_AttackRate = float(Control_CWOP_AttackRate) / cwopCount
    
    print "Theoretical strongest and weakest pharmacuetical intervention (PI) & logistical implementation (LI):"
    print "\nActive Duty Military on base:", admCount
    print "Civilian Workers on post", cwopCount
    print "\n\t\tActive Duty\tCivilian Workers"
    print "Strong PI, Strong LI\t", percent(SISE_ADM_Percent), '\t', percent(SISE_CWOP_Percent)
    print "Weak PI, Strong LI\t", percent(SIWE_ADM_Percent), '\t', percent(SIWE_CWOP_Percent)
    print "Strong PI, Weak LI\t", percent(WISE_ADM_Percent), '\t', percent(WISE_CWOP_Percent)
    print "Weak PI, Weak LI\t", percent(WIWE_ADM_Percent), '\t', percent(WIWE_CWOP_Percent)
    print "Best Intervention\t", percent(Best_ADM_Percent), '\t', percent(WIWE_CWOP_Percent)
    print "Worst Intervention\t", percent(Worst_ADM_Percent), '\t', percent(WIWE_CWOP_Percent)
    print "No Intervention\t\t", percent(Control_ADM_AttackRate), '\t', percent(WIWE_CWOP_Percent)
    
    return 'null'
    
```

#### pullExperiment: Main Cell Execution Function, pulls and passes data sequentially through analysis functions

```python
def pullExperiment(expDir,end):
    seeded = str(getSeedNumber(expDir))
    control = 'seeds' + seeded
    print seeded, "Infections Seeded\n"
    
    admExperimentStats = getExperiment(expDir+admStats)
    cwopExperimentStats = getExperiment(expDir+cwopStats)
    correlationData = getCorrelations(admExperimentStats['iterations'],natural,experimental)
    maximaData = getMaxima(admExperimentStats['means'])
    curves = getCurves(expDir,maximaData['weakRef'],maximaData['strongRef'],admFile,end,control)
    ranks = getRankedInterventions(admExperimentStats['means'],cwopExperimentStats['means'],maximaData['strongRef'],maximaData['weakRef'],control)
    print '\n\n'
    
    
```
